```python
import requests
import pandas as pd
import re

""" 
def extract_ids(url):
    
    # Extract numbers from the URL
    numbers = re.findall(r'\d+', url)
    
    # Assuming the URL structure is consistent
    # Get the rightmost number as item_id and the second to the rightmost number as shop_id
    if len(numbers) >= 2:
        item_id = numbers[-1]
        shop_id = numbers[-2]
        return shop_id, item_id
    else:
        return None, None
 """

def extract_ids(url):
    # Regular expression pattern to extract numbers between "i" and "?sp_atk"
    pattern = r'i\.(\d+)\.(\d+)\?sp_atk'
    
    # Find the match using the pattern
    match = re.search(pattern, url)
    
    if match:
        shop_id = match.group(1)
        item_id = match.group(2)
        return shop_id, item_id
    else:
        return None, None

# ---------------------------------------------------------------------- PASTE PRODUCT URL HERE
shop_url = 'https://shopee.ph/Baseus-WM02-True-Wireless-Earphones-i.131196305.17552845260?sp_atk=ef0e9936-ea9d-41f0-8ad6-b0470408a11e&xptdk=ef0e9936-ea9d-41f0-8ad6-b0470408a11e'
shop_id, item_id = extract_ids(shop_url)



url = "https://shopee.com.my/api/v2/item/get_ratings"           # API

limit = 4                       # What happens here is basically, gakwa sha limit of 4. Tapos sa dalom may ara command nga offset += limit
offset = 0                      # This means na ang starting review is now 4. Ang offset is the one indicator kung diin masugod kwa review, and it is being updated constantly.
reviews = []








# Initializes a session object from the requests module, which allows for persistent sessions. This can be beneficial for making multiple requests to the same server, as it keeps certain parameters (like cookies) between requests.
session = requests.Session()     

# Updates the HTTP headers of the session object. In this case, it seems to be setting some cookies in the header. Cookies are typically used for maintaining stateful information across HTTP requests.
session.headers.update({"Cookie": "_gcl_au=xxxx; SPC_IA=-1; SPC_EC=-; SPC_F=xxxx; SPC_U=-; SPC_T_ID=xxxx; SPC_T_IV=xxxx; SPC_SI=xxxx; _ga=xxxx; _gid=xxxx; cto_lwid=xxxx; _fbp=xxxx; _hjid=xxxx; SPC_SIxxxx=xxxx"})

while True:

    # A dictionary params is constructed with parameters for an HTTP GET request. These parameters seem to be related to querying an API, including item ID, shop ID, offset, limit, and various filtering options.
    params = {
        "itemid": item_id,
        "shopid": shop_id,
        "offset": offset,
        "limit": limit,
        "filter": "0",
        "flag": "0",
        "sort": "0",
        "append": "0",
        "before_bundle": "",
        "language": "en",
    }

    # Sends an HTTP GET request to a URL specified by url, passing the parameters params. The response from the server is stored in the response variable, and .json() method converts the response to a JSON object.
    response = session.get(url, params=params).json()

    if response["error"]:
        print(f"Error: {response['error']}")
        break

    elif response["data"]["ratings"]:
        print(offset)
        print (response["data"]["ratings"])

        for rating in response["data"]["ratings"]:

            # Replace newline characters with space in the comment
            comment = rating.get('comment', '').replace('\n', ' ')
            rating['comment'] = comment

            reviews.append(rating)
            offset += limit

            # time.sleep(1)  # introduce a delay  of 1 second between requests

    else:
        break














df = pd.DataFrame(reviews)

print(f"Retrieved {len(reviews)} reviews.")

# select columns to keep in the final CSV file
# df = df[['orderid', 'itemid', 'cmtid', 'ctime', 'rating', 'userid', 'shopid', 'comment', 'rating_star', 'model_name']]

df = df[['shopid', 'itemid', 'userid', 'rating_star', 'comment']]
# df = df.to_string(index=False)

print(df.columns)

# save the DataFrame as a CSV file
df.to_csv('scraped_reviews_test.csv', index=False)


# Previous shop_id & item_id used:
# shop_id = "592249106"         item_id = "15347304488"         url = https://shopee.ph/Extended-Mousepad-Gaming-High-Quality-Exquisite-Pattern-Designs-800*300*3MM-Gaming-Mouse-Pad-Large-i.592249106.15347304488?sp_atk=1ccd924d-9b88-4d81-b110-3c584f508f9d&xptdk=1ccd924d-9b88-4d81-b110-3c584f508f9d

# shop_id = "131196305"         item_id = "17552845260"         url = https://shopee.ph/product/131196305/17552845260
# shop_id = "282024671"         item_id = "23576145769"         url = https://shopee.ph/product/282024671/23576145769
# shop_id = "153457585"         item_id = "2463825120"          url = https://shopee.ph/product/153457585/2463825120
# shop_id = "339409991"         item_id = "8964357378"          url = https://shopee.ph/product/339409991/8964357378
# shop_id = "426990928"         item_id = "7187083641"          url = https://shopee.ph/product/426990928/7187083641
```
